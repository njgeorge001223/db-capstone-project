```python
CONSTANT_LOCAL_FILE_NAME = 'db_client.ipynb'
```

### **Part 1: Open MySQL Database Connection**

```python
connection \
    = connector.connect \
        (user = db_user_name, 
         password = db_user_password, 
         db = db_name)

cursor = connection.cursor()
```

### **Part 2: Display Database Tables**

```python
query = 'SHOW tables'

cursor.execute(query)
```

```python
results = cursor.fetchall()

for row in results: print(row[0])
```

### **Part 3: Use JOIN Query to List Customers and Money Spent**

```python
query \
    = """SELECT customers.full_name, customers.contact_number, orders.total_cost
         FROM customers
         INNER JOIN bookings ON customers.customer_id = bookings.customer_id
         INNER JOIN orders ON bookings.booking_id = orders.booking_id
         WHERE orders.total_cost > 60;"""

cursor.execute(query)
```

```python
results = cursor.fetchall()

print('--- Promotional Campaign List (Orders > $60) ---')

for row in results:
    
    print(f"Customer: {row['full_name']} | Contact: {row['contact_number']} | Spent: ${row['total_cost']}\n")
```

### **Part 4: Close Database Connection**

```python
if 'connection' in locals() and connection.is_connected():
    
    cursor.close()
    
    connection.close()
    
    print("The script closed the MySQL connection.")
```
